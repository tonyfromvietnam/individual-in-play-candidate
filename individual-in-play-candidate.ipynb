## Re-structure the data, changing their fields

```python
# Rename the Kpi Date columns to CV Sent Date, change that to datetime type
cv_sents.rename(columns={'Created Date' : 'CV Sent Date', '﻿"Job ID"' : 'Job ID'}, inplace=True)

# Change the CV Sent Date to datetime type
cv_sents['CV Sent Date'] = pd.to_datetime(cv_sents['CV Sent Date'], format='%d/%m/%Y')
cv_sents = cv_sents[['Job ID', 'CV Sent Date', 'Owner', 'Company', 'Candidate']]

# Display the new dataset
cv_sents.head()
```

```python
# Rename the Kpi Date columns to Interview Date, change that to datetime type
interviews.rename(columns={'Created Date' : 'Interview Date'}, inplace=True)

# Change the Create Date to datetime type
interviews['Interview Date'] = pd.to_datetime(interviews['Interview Date'], format='%d/%m/%Y')
interviews = interviews[['Job ID', 'Interview Date', 'Owner', 'Company', 'Candidate']]

# Display the new dataset
interviews.head()
```

```python
# Rename the Kpi Date columns to Placements Date, change that to datetime type
placements.rename(columns={'Kpi Date' : 'Placement Date'}, inplace=True)

# Change the Create Date to datetime type
placements['Placement Date'] = pd.to_datetime(placements['Placement Date'], format='%d/%m/%Y')

# Display the new dataset
placements.head()
```

## Create a few helper columns to get the status of CVs Sent to Clients

```python
# Apply helper columns to get the status of CVs
interviews['Interview'] = True
rejects['Rejected'] = True
```

```python
# Merge the CV record with the interview record
merged_cv_sent_status = cv_sents.merge(
    interviews,
    on=['Company','Job ID', 'Owner', 'Candidate'],
    how='left'
)


# Merge the CV record with the rejected record
merged_cv_sent_status = merged_cv_sent_status.merge(
    rejects,
    on=['Company', 'Job ID', 'Owner', 'Candidate'],
    how='left'
)

# Print the final list
display(merged_cv_sent_status)
```

```python
# Create the conditions
conditions = [
    merged_cv_sent_status['Interview'].notna() & merged_cv_sent_status ['Rejected'].notna(),
    merged_cv_sent_status['Interview'].notna(),
    merged_cv_sent_status['Rejected'].notna()
]

# Create the choices (Interview = Converted, "Rejected = "Non-converted")
choices = [
    "Converted",
    "Converted",
    "Non-converted"
]

merged_cv_sent_status["Status"] = np.select(conditions, choices, default="N/A")

display(merged_cv_sent_status)
```

```python
# Drop the support columns
merged_cv_sent_status = merged_cv_sent_status.drop(columns = ['Interview', 'Rejected'])
```

```python
# Drop the N/A rows in the status
merged_cv_sent_status = merged_cv_sent_status[(merged_cv_sent_status.Status != 'N/A')]

# Drop the Interview Date < CV Sent Date records
merged_cv_sent_status = merged_cv_sent_status[(merged_cv_sent_status['CV Sent Date'] <= merged_cv_sent_status['Interview Date']) | (merged_cv_sent_status['Status'] == 'Non-converted')]

display(merged_cv_sent_status)
```

```python
# Adding the column tracking the number of days convert from CV Sent to 1st Interview

merged_cv_sent_status['Time to Convert (Days)'] = (merged_cv_sent_status['Interview Date'] - merged_cv_sent_status['CV Sent Date']).dt.days
```
