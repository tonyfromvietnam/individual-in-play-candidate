## Re-structure the data, changing their fields

```python
# Rename the Kpi Date columns to CV Sent Date, change that to datetime type
cv_sents.rename(columns={'Created Date' : 'CV Sent Date', '﻿"Job ID"' : 'Job ID'}, inplace=True)

# Change the CV Sent Date to datetime type
cv_sents['CV Sent Date'] = pd.to_datetime(cv_sents['CV Sent Date'], format='%d/%m/%Y')
cv_sents = cv_sents[['Job ID', 'CV Sent Date', 'Owner', 'Company', 'Job Title', 'Candidate']]

# Display the new dataset
cv_sents.head()
```

```python
# Rename the Kpi Date columns to Interview Date, change that to datetime type
first_interviews.rename(columns={'Interview Date' : '1st_Interview Date'}, inplace=True)
second_interviews.rename(columns={'Interview Date' : '2nd_Interview Date'}, inplace=True)
third_interviews.rename(columns={'Interview Date' : '3rd_Interview Date'}, inplace=True)
fourth_interviews.rename(columns={'Interview Date' : '4th_Interview Date'}, inplace=True)

# Change the Create Date to datetime type
first_interviews['1st_Interview Date'] = pd.to_datetime(first_interviews['1st_Interview Date'], format='%d/%m/%Y')
first_interviews = first_interviews[['Job ID', 'Job Title', '1st_Interview Date', 'Owner', 'Company', 'Candidate']]

second_interviews['2nd_Interview Date'] = pd.to_datetime(second_interviews['2nd_Interview Date'], format='%d/%m/%Y')
second_interviews = second_interviews[['Job ID', 'Job Title', '2nd_Interview Date', 'Owner', 'Company', 'Candidate']]

third_interviews['3rd_Interview Date'] = pd.to_datetime(third_interviews['3rd_Interview Date'], format='%d/%m/%Y')
third_interviews = third_interviews[['Job ID', 'Job Title', '3rd_Interview Date', 'Owner', 'Company', 'Candidate']]

fourth_interviews['4th_Interview Date'] = pd.to_datetime(fourth_interviews['4th_Interview Date'], format='%d/%m/%Y')
fourth_interviews = fourth_interviews[['Job ID', 'Job Title', '4th_Interview Date', 'Owner', 'Company', 'Candidate']]
```

```python
# Rename the Kpi Date columns to Offer Date, change that to datetime type
offers.rename(columns={'﻿"Created Date"' : 'Offer Date'}, inplace=True)

# Change the Create Date to datetime type
offers['Placement Date'] = pd.to_datetime(offers['Offer Date'], format='%d/%m/%Y')

# Change the columns for offers
offers = offers[['Job ID', 'Job Title', 'Owner', 'Company', 'Candidate']]

# Display the new dataset
offers.head()
```

## Create a few helper columns to get the status of CVs Sent to Clients

```python
# Apply helper columns to get the status of CVs
first_interviews['Interview_1st'] = True
second_interviews['Interview_2nd'] = True
third_interviews['Interview_3rd'] = True
fourth_interviews['Interview_4th'] = True
offers['Offer'] = True
rejects['Rejected'] = True
```

```python
first_interviews.head()
```

### Merge the CV records with the interviews

```python
# Merge the CV record with the 1st interview record
merged_cv_sent_status = cv_sents.merge(
    first_interviews,
    on=['Company','Job ID', 'Owner', 'Candidate', 'Job Title'],
    how='outer'
)

# Merge the CV record with the 2nd interview record
merged_cv_sent_status = merged_cv_sent_status.merge(
    second_interviews,
    on=['Company','Job ID', 'Owner', 'Candidate', 'Job Title'],
    how='outer'
)

# Merge the CV record with the 3rd interview record
merged_cv_sent_status = merged_cv_sent_status.merge(
    third_interviews,
    on=['Company','Job ID', 'Owner', 'Candidate', 'Job Title'],
    how='outer'
)

# Merge the CV record with the 4th interview record
merged_cv_sent_status = merged_cv_sent_status.merge(
    fourth_interviews,
    on=['Company','Job ID', 'Owner', 'Candidate', 'Job Title'],
    how='outer'
)

# Merge the CV record with the offers record
merged_cv_sent_status = merged_cv_sent_status.merge(
    offers,
    on=['Company','Job ID', 'Owner', 'Candidate', 'Job Title'],
    how='outer'
)

# Merge the CV record with the rejected record
merged_cv_sent_status = merged_cv_sent_status.merge(
    rejects,
    on=['Company', 'Job ID', 'Owner', 'Candidate', 'Job Title'],
    how='outer'
)

# Print the final list
display(merged_cv_sent_status)
```

### Change the status of the candidates

```python
# Create the conditions
conditions = [
    merged_cv_sent_status['Rejected'].notna(),
    merged_cv_sent_status['Interview_4th'].notna() & merged_cv_sent_status ['Rejected'].isna() & merged_cv_sent_status ['Offer'].isna(),
    merged_cv_sent_status['Interview_3rd'].notna() & merged_cv_sent_status ['Rejected'].isna() & merged_cv_sent_status ['Interview_4th'].isna(),
    merged_cv_sent_status['Interview_2nd'].notna() & merged_cv_sent_status ['Rejected'].isna() & merged_cv_sent_status ['Interview_3rd'].isna(),
    merged_cv_sent_status['Interview_1st'].notna() & merged_cv_sent_status ['Rejected'].isna() & merged_cv_sent_status ['Interview_2nd'].isna(),
    merged_cv_sent_status['Rejected'].isna() & merged_cv_sent_status['Interview_1st'].isna() & merged_cv_sent_status ['Offer'].isna()
]

# Create the choices (Interview = Converted, "Rejected = "Non-converted")
choices = [
    "Rejected",
    "4th Interview",
    "3rd Interview",
    "2nd Interview",
    "1st Interview",
    "CV Sent"
]

merged_cv_sent_status["Status"] = np.select(conditions, choices, default="N/A")

display(merged_cv_sent_status)
```

### Add the column of Action Date - Interview/CV Sent Date

```python
def get_action_date(row):
    status = row['Status']
    
    # Map status to the corresponding column name
    if status == 'CV Sent':
        return row['CV Sent Date']
    elif status == '1st Interview':
        return row['1st_Interview Date']
    elif status == '2nd Interview':
        return row['2nd_Interview Date']
    elif status == '3rd Interview':
        return row['3rd_Interview Date']
    elif status == '4th Interview':
        return row['4th_Interview Date']
    else:
        return pd.NaT # Return "Not a Time" for unknown statuses

# --- Apply the Function ---
merged_cv_sent_status['Action Date'] = merged_cv_sent_status.apply(get_action_date, axis=1)
```

```python
# Drop the N/A rows in the status
merged_cv_sent_status = merged_cv_sent_status[(merged_cv_sent_status.Status != 'N/A') & (merged_cv_sent_status.Status != 'Rejected')]

merged_cv_sent_status['Status'].value_counts()
```

```python
# Drop the support columns
merged_cv_sent_status = merged_cv_sent_status.drop(columns = ['Interview_1st', 'Interview_2nd', 'Interview_3rd', 'Interview_4th', 'Rejected', 'Offer', 
                                                              '1st_Interview Date', '2nd_Interview Date', '3rd_Interview Date', '4th_Interview Date'])
```

## Export the in-play candidates for the recruiters

```python
owners_list = merged_cv_sent_status['Owner'].dropna().unique()

merged_cv_sent_status['Action Date'].dt.floor('d')

# Create a folder to keep things organised
output_folder = 'Consultant_Active_Reports'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
```

```python
for owner in owners_list:
    # 1. Filter for this specific owner
    owner_data = merged_cv_sent_status[merged_cv_sent_status['Owner'] == owner].copy()
    
    # 2. Sort by Job ID (Ascending)
    owner_data = owner_data.sort_values(by='Job ID', ascending=True)
    
    # 3. Select clean columns for the consultant
    export_cols = ['Job ID', 'Company', 'Job Title', 'Candidate', 'Status']
    final_export = owner_data[export_cols]
    
    # 4. Define Filename (Clean the name to avoid file errors)
    clean_owner_name = str(owner).replace(" ", "_").replace("/", "-")
    filename = f"{output_folder}/Active_Candidates_{clean_owner_name}.csv"
    
    # 5. Export to csv
    final_export.to_csv(filename, index=False)
    
    print(f"Exported: {filename} ({len(final_export)} candidates)")

print("\n--- ALL EXPORTS COMPLETE ---")
```
